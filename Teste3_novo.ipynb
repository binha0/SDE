```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# definindo parametros 
T = 1
m = 10000
n = 1000

ts = np.array([j*T / n for j in range(n+1)]) # len 101 n+1
```

```python
# calculo da integral 
Is = []
for i in range(m):
    Ws = np.array([np.random.normal(0,np.sqrt(ts[j])) for j in range(n+1)])
    dW = Ws[1:n+1] - Ws[0:n]
    Is.append(sum(Ws[0:n] * dW))
```

```python
Is = []
for i in range(m):
    aux = []
    for j in range(n+1):
        aux.append(np.sqrt(j*(j+1)*np.random.gamma(0.5,2*T/n) - np.random.gamma(0.5,2*ts[j])))
    Is.append(sum(np.array(aux)))
```

```python
# calculando aproximações gamma
gamma = 0.5 * (np.random.gamma(0.5,2*T,size=m) - T)
aprox = 0.5 * (np.random.normal(0,T,size=m)**2 - T)
```

```python
plt.hist(Is,color='crimson',label='discretização')
plt.grid()
plt.legend()
```

```python
#plt.hist(Is)
plt.hist(gamma,alpha=0.5, label="Aproximação Gamma")
plt.hist(aprox,alpha=0.5,label="Solução Analítica")
plt.legend()
plt.grid()
```
